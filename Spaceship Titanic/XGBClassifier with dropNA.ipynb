## XGBoost

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = train.drop('Transported', axis=1)
y = train['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# XGB
xgb = XGBClassifier(n_estimators=100, max_depth=4, random_state=2)

# Fit XGB to traning set
xgb.fit(X_train, y_train)

# Predict labels of test set, y_pred
y_pred = xgb.predict(X_test)

score = accuracy_score(y_pred, y_test)

print('Score: ' + str(score*100))
```

```python
# Initialize XGB
xgb = XGBClassifier(booster='gbtree', max_depth=4, learning_rate=0.1, n_estimators=100, random_state=2, n_jobs=-1)

# Fit XGB to traning set
xgb.fit(X_train, y_train)

# Predict labels of test set, y_pred
y_pred = xgb.predict(X_test)

score = accuracy_score(y_pred, y_test)

print('Score: ' + str(score*100))
```
